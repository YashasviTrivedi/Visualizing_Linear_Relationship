```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/Instagram.csv", encoding='cp1252')
data=data.dropna()
print(data.head())
```

```python
figure = px.scatter(data_frame = data, x= 'Impressions', y='Likes', size='Likes', trendline='ols',title='Relationship between Likes and Impressions')
figure.show()
```

```python
plt.figure(figsize=(10,8))
plt.style.use('fivethirtyeight')
plt.title('Relationship Between Likes and Impressions')
sns.regplot(x='Impressions',y='Likes',data=data)
plt.show()
```
